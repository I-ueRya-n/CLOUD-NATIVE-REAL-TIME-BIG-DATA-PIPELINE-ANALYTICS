```python
from atproto import Client
import dotenv
import os

dotenv.load_dotenv()
bluesky_username = os.getenv('BLUESKY_USERNAME')
bluesky_password = os.getenv('BLUESKY_PASSWORD')
client = Client()
client.login(bluesky_username, bluesky_password)

cursor = ''
data = client.app.bsky.feed.search_posts({
  "q": "Pauline Hanson",
  "sort": "latest",
  'limit': 3,
}, headers={'Accept-Language': "en"})
posts = data.posts
for post in posts:
    # print(post)
    if post.author:
        print("Post by:", post.author.display_name)
    
    record = post.record
    print("Post at:", record.created_at)
    print("Post text:", record.text)
    if record.embed and record.embed.py_type == "app.bsky.embed.images":
        for image in record.embed.images:
            print("Image alt text:", image.alt)
    print("-----")
# print(data)
```
